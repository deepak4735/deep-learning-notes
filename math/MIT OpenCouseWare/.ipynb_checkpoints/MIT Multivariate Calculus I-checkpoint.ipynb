## Vectors and Matrices
### Law of Cosines
$$|\vec{c}|^2 = |\vec{a}|^2 + |\vec{b}|^2 - |\vec{a}||\vec{b}| \cos(\theta)$$
 
The sign of the dot product is the same as the sign of cos(theta). If the angles is > 90 the sign will be negative, zeor at 90, and postisitve at < 90 

### Detect Orthogonality 
If the dot product of two vectors is zero, they are perpendicular. 

### Finding components
Finding the component of vector A along unit vector u.
$$|\vec{a}| cos(\theta ) = \vec{a} \cdot \hat{u}$$

### Cross Product
$$ \vec{a} x \vec{b} = \Bigg[ \matrix{\hat{i} &\hat{j} &\hat{k} \cr a_1 &a_2 &a_3 \cr b_1 &b_2 &b_3} \Bigg] $$

The length of the cross product is equal to the area of the parallelogram created by the cross-product of A and B. 

The cross product, unlike the dot product results in a vector not a scalar, and has a firtst row of unit vectors. The dir of A x B is perpendicular to the plane of the parallelogram formed by A and B's determinant. 

Use the right hand rule to show where A x B will point, the vector formed by their cross product will point in the direction of your thumb. 

A x B = -B x A

Swapping the order swaps the directions

A x A = 0

### Application of Cross Product
If we have three vectos and we want to see if they form a plane- then we can look at the determinant of the three vectors and if it is zero we have a plane. The vector perpidicular to the plane is called the normal vector. 

$$ det(\vec{p_1 p},\ \vec{p_1 p_2},\ \vec{p_1 p_3}) = 0$$

One of the most important uses of the cross product will be finding  the normal vector to a plane given two vectors.  

### Matrices
Construction of new matrices, m x n rows and columns. If AB, the width of A has to be equal to the heigth of B (or equal to 1 and use broadcasting).

Matrix products are associative, not cross product though. 
$$(AB)X = A(BX)$$
$$AB \neq BA$$

**Identiy matrix**
$$IX = X$$
The identity is composed with 1's along the diagonal. 

**Rotation in the plane by 90 degrees counter-clockwise**
$$\Bigg[ \matrix{0 &-1 \cr 1 &0} \Bigg]$$

**Rotation in the plane by 180 degrees counter-clockwise, the negative identity matrix**
$$\Bigg[ \matrix{-1 &0 \cr 0 &-1} \Bigg]$$

**Inverse Matrix**
$$M = A^{-1}$$
$$AM = I$$
$$MA = I$$

Formula for small matrices, for small matrices:
Calulate the adjoint matrix and divide by the determinant
$$A^{-1} = \frac{adjoint(A)}{determinant(A)}$$

### Equations of Planes
How they relate to linear systems and matrices. How do we define a plane given a normal vector. Given a normal vector and a plane defined at P_0, we can construct a plane equation as $x + 5Y + 10z = -3$. Where we use the coordinates of the normal vector to construct the equation for the plane and we use the point on the plane plugged into that equation to give us the result, in this case (-3). We can now describe the plane. 

You could also get the normal vector by taking the cross-product of two vectors in the plane. 

Testing if vector is parallel, perpindicular, or neither to plane. We can extract the normal vector to the plane with the plane's equation and test if is scales to the test vector because all normal vectors are parallel. If the normal vector dot product to the test vector equals zero then it is perpendicular. 

A normal vector to the plane can be extracted from the coefficiants of the plane's equation.

### Three Intersecting Planes
A solution to 3 intersecting planes can be (a) a single point (b) a line (c) or no solution (d) a plane- in the case of all three planes parallel. 

Matrix A is invertible when the determinant is *not* 0. We divide by the determinant as the last step and if the determinant is zero there is no inverse. 

The **homogenous case** is where AX = 0. You can have a system of equations where they all equal zero. 

Given three planes, if N_1 and N_2 are perpendicular then so is N_3.

**General case** if AX = B. If $det(A) \neq 0$ then a unique solution exists, $X = A^{-1}B$. If the det(A) = 0, then we have no solutions or infinitely many. 

### Equations of Lines
Think of a line as the trajectory of a moving point. This is called a *parametric equation* of lines.
